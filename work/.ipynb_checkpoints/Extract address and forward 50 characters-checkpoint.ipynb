```python
data_dir = "../data/"
src_dir = "../src/"

lc_fname = "listed_company"
pr_fname = "test.csv"
#pr_fname = "pressrelease_all.csv"
#pr_info_fname = "pressrelease_info.csv"
```

```python
lc_data = []
with open(data_dir + lc_fname, 'r') as f:
    for line in f:
        lc_data.append(tuple(line.strip().split('\t')))
        
print(lc_data[:5])
```

```python
pr_data = []
with open(data_dir + pr_fname, 'r') as f:
    for line in f:
        pr_data.append(tuple(line.strip().split('\t')))
        
print(pr_data[:3])
```

```python
import subprocess

pr_address = {}
for pr in pr_data[:10]:
    pr_bodysub = pr[2]
    with open("tmp.bodysub", 'w') as f:
        f.write(pr_bodysub)
    ret = subprocess.getoutput("./" + src_dir + "extract_address.pl tmp.bodysub").strip().split('\n')
    pr_address[pr[0]] = set(ret[2:])
```

```python
import re
from collections import defaultdict

offset_characters = 30   # 住所の位置から何文字前までの範囲で企業名を探すか

for pr in pr_data[:10]:
    #print(pr[2])
    pr_address_set = pr_address[pr[0]]
    print(pr_address_set)
    for address in pr_address_set:
        pat = re.compile(r'%s' % address)
        
        for match in pat.finditer(pr[2]):
            start_pos = max([match.start() - offset_characters, 0])
            #print(match.start())
            pr_bodysub_part = pr[2][start_pos: match.start()]
            # pr_bodysub_part から(住所も一致する)企業名を探し出す
            find_match(pr_bodysub_part, address)   # 下に定義
```
